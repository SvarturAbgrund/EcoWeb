````python
# FMEA/AMEF Risk Management for EcoWeb Project

This notebook analyses the entire EcoWeb project and produces a structured risk management template.  The resulting data can be exported to an Excel workbook named **FMEA – EcoWeb.xlsx**.

---

## 1. Project Overview and Structure Analysis

The EcoWeb application consists of a PHP-based MVC‑like structure with:

- **Data layer** (`data/`): authentication (`auth.php`), models for comments and users, database connection, session security.
- **Presentation layer** (`presentation/`): pages for login, registration, ecosystems, comments, etc. Includes a shared header layout.
- **Public assets** (`public/`): CSS, JavaScript and images for the front end.

**Project name:** EcoWeb
**Team members:** *[Complete with your names]*

| Component folder | Purpose |
|------------------|---------|
| data             | Business logic, DB access, security functions |
| presentation     | User‑facing pages and layout |
| public           | Static assets (styles, scripts, images) |

---

## 2. Risk Identification by Process

We consider the following processes/functions:

- Análisis
- Diseño
- Codificación
- Pruebas
- Implementación
- Documentación
- Administración de proyectos

For each process we list potential failure modes (riesgos):

- **Análisis:** Requisitos incompletos, mal entendimiento del dominio.
- **Diseño:** Arquitectura inadecuada, falta de modularidad.
- **Codificación:** Vulnerabilidades de seguridad, errores de lógica, mala gestión de sesiones.
- **Pruebas:** Cobertura insuficiente, datos de prueba inadecuados.
- **Implementación:** Configuración de servidor incorrecta, pérdida de datos.
- **Documentación:** Documentación desactualizada o inexistente.
- **Administración de proyectos:** Cronograma irreal, comunicación deficiente.

---

## 3. Define Probability and Impact Scales

**Ocurrencia (Probabilidad)**

| Nivel | Descripción |
|-------|-------------|
| 1     | Muy baja (rara vez ocurre) |
| 3     | Baja |
| 5     | Moderada |
| 7     | Alta |
| 9     | Muy alta (ocurre frecuentemente) |

**Severidad (Impacto)**

| Nivel | Descripción |
|-------|-------------|
| 1     | Insignificante |
| 3     | Menor |
| 5     | Moderado |
| 7     | Grave |
| 9     | Catastrófico |

Detectabilidad se evalúa de forma similar (1 = fácil de detectar, 10 = casi imposible).

---

## 4. Qualitative Risk Analysis

A sample table of identified risks with qualitative attributes:

| Proceso        | Modo de Falla                  | Efecto Potencial                  | Severidad | Causa Potencial                       | Ocurrencia | Controles Actuales                | Detectabilidad |
|----------------|--------------------------------|------------------------------------|-----------|----------------------------------------|------------|------------------------------------|----------------|
| Codificación   | Inyección SQL                 | Acceso no autorizado, corrupción DB | 9         | Validación insuficiente de entradas   | 7          | `mysqli` con parámetros, seguridad | 5              |
| Pruebas        | Cobertura insuficiente        | Bugs en producción                 | 7         | Falta de recursos de prueba           | 5          | Pruebas manuales básicas           | 4              |
| Implementación | Config. servidor incorrecta   | Sitio inaccesible                  | 8         | Desconocimiento de ambiente produ.    | 5          | Checklist de despliegue            | 6              |

(El equipo puede ampliar esta tabla con más riesgos y ajustar valores.)

---

## 5. Quantitative Risk Analysis

Calcular el **Número de Prioridad de Riesgo (RPN)**: `RPN = Severidad × Ocurrencia × Detectabilidad`.

| Proceso      | Modo de Falla       | S | O | D | RPN |
|--------------|---------------------|---|---|---|-----|
| Codificación | Inyección SQL       | 9 | 7 | 5 | 315 |
| Pruebas      | Cobertura insuf.    | 7 | 5 | 4 | 140 |
| Implement.   | Config. errónea     | 8 | 5 | 6 | 240 |

Ordenar por RPN para priorizar las acciones.

---

## 6. Risk Response Planning

Para cada riesgo definimos acciones y responsables:

| Riesgo                | Acciones recomendadas                          | Responsable        |
|-----------------------|------------------------------------------------|--------------------|
| Inyección SQL         | Utilizar consultas preparadas, revisar inputs  | Equipo de desarrollo |
| Cobertura insuf.      | Añadir pruebas automatizadas, revisar casos    | QA / Tester        |
| Config. servidor err. | Crear script de despliegue, documentar pasos   | DevOps / Admin     |


---

## 7. Generate FMEA Report Summary

Creamos un DataFrame con todas las columnas anteriores, listo para exportar a Excel. La hoja de portada contendrá el nombre del proyecto y miembros.

```python
import pandas as pd

# ejemplo de tabla
risks = [
    {"Proceso":"Codificación","Modo":"Inyección SQL","Severidad":9,"Ocurrencia":7,"Detectabilidad":5,"RPN":315},
    {"Proceso":"Pruebas","Modo":"Cobertura insuficiente","Severidad":7,"Ocurrencia":5,"Detectabilidad":4,"RPN":140},
]

df = pd.DataFrame(risks)

with pd.ExcelWriter('c:/xampp/htdocs/EcoWeb/EcoWeb/FMEA – EcoWeb.xlsx') as writer:
    # portada
    cover = pd.DataFrame({"Proyecto":["EcoWeb"],"Integrantes":["Nombres del equipo..."]})
    cover.to_excel(writer, sheet_name='Portada', index=False)
    df.to_excel(writer, sheet_name='Tabla de riesgos', index=False)
```

Al ejecutar el código anterior se generará el archivo Excel solicitado.

---

🎯 **Resultado**: el análisis de riesgos ya está documentado y estructurado. El equipo puede editar los valores en la hoja `Tabla de riesgos` del archivo `FMEA – EcoWeb.xlsx`, agregar nuevas filas o ajustar escalas según lo consideren apropiado.

¡Listo! Con esto tienes un esquema completo para la gestión de riesgos en formato FMEA/AMEF para tu proyecto EcoWeb. Te toca ampliar y personalizar con tu equipo.
````
